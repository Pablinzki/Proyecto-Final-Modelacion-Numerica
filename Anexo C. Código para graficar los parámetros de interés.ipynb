Importar paqueterías

```python
import MITgcmutils as MIT
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

Creación de los ejes para los arreglos

```python
depth = MIT.mds.rdmds('Depth')
z = MIT.mds.rdmds('RC')
z = np.reshape(z, (15))
y = MIT.mds.rdmds('YC')
x = MIT.mds.rdmds('XC')
```

Creación de la máscara de tierra

```python
# Máscara de tierra - para quitar las celdas que son tierra y quedarnos solo con el agua
# obtenemos la máscara (arreglo de booleanos donde tenemos True donde hay 
# tierra y False donde hay agua, en superficie)
hFacC = MIT.mds.rdmds('hFacC')

hFacC_mask = np.ma.masked_less(hFacC, 1)

mask = np.ma.getmask(hFacC_mask)

mask
```

## Gráficas de los parámetros físicos

Profundidad

```python
H = MIT.mds.rdmds('Depth')
Hmasked = np.ma.masked_array(H, mask = mask[0,:,:])
plt.contourf(x, y, Hmasked)
plt.colorbar()
plt.xlabel('lon')
plt.ylabel('lat')
plt.title('Profundidad del fluido (m)')
```

Altura de referencia de la superficie oceánica

```python
Eta = MIT.mds.rdmds('Eta.0005184000')
Eta_masked = np.ma.masked_array(Eta, mask = mask[0,:,:])
plt.contourf(x, y, Eta_masked)
plt.colorbar()
plt.xlabel('lon')
plt.ylabel('lat')
plt.title('Altura de referencia de la superficie oceánica')
```
